```python
import pandas as pd
import numpy as np

#import csv file into a pandas dataframe
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url, header=None)
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
# add column names to the file
df.columns = headers
```

```python
# print first five rows
df.head(5)
```

```python
# missing data comes with the question mark "?". We replace "?" with NaN (Not a Number), which is Python's default missing value marker, 
# for reasons of computational speed and convenience
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# the missing values are converted to default and then counted using for loop.
# True" stands for missing value, while "False" stands for not missing value.
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")   
```

```python
# replace missing values with average for the columns with missing values as seen in previous output except price
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of bore:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
#idxmax()" method to calculate the most common type automatically:
df['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent 
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# simply drop whole row with NaN in "price" column as we are predicting the column price
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
# now we have dataset with no missing values.now Convert data types to proper format
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

Now, we finally obtain the cleaned dataset with no missing values and all data in its proper format.Now e do data standardization.

```python
# we do standardization like changing variable to common standard mpg to liters per km
# Convert mpg to L/100km by mathematical operation (235 divided by mpg)
df['city-L/100km'] = 235/df["city-mpg"]

# check your transformed data 
df.head()
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={'"highway-mpg"':'highway-L/100km'}, inplace=True)

# check your transformed data 
df.head()
```

```python
# transform mpg to L/100km by mathematical operation (235 divided by mpg)
df["highway-mpg"] = 235/df["highway-mpg"]

# rename column name from "highway-mpg" to "highway-L/100km"
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)

# check your transformed data 
df.head()
```

```python
# data normalization.transforming values of several variables into a similar range
# replace (original value) by (original value)/(maximum value)
# include scaling the variable so the variable average is 0, scaling the variable so the variance is 1, 
# or scaling variable so the variable values range from 0 to 
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
```
